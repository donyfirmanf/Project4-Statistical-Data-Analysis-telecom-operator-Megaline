## Deskripsi Paket Prabayar

Megaline membulatkan detik ke menit, serta membulatkan megabita ke gigabita. Untuk panggilan, setiap panggilan individual dibulatkan ke atas: bahkan jika panggilan tersebut berlangsung hanya satu detik, ia akan dihitung sebagai satu menit. Untuk lalu lintas web, setiap sesi web tidak dibulatkan ke atas. Akan tetapi, total untuk sebulan dibulatkan ke atas. Jika seorang pengguna menghabiskan 1025 megabita bulan ini, maka ia akan dikenai biaya untuk 2 gigabita.

**Surf:**
- Biaya bulanan: USD20
- 500 menit durasi panggilan per bulan, 50 SMS, dan 15 GB data
- Setelah melebihi batas paket, akan dikenakan:
        1 menit: 3 sen
        1 SMS: 3 sen
        1 GB data: USD10

**Ultimate:**
- Biaya bulanan: USD70
- 3000 menit durasi panggilan per bulan, 1000 SMS, dan 30 GB data
- Setelah melebihi batas paket, akan dikenakan:
        1 menit: 1 sen
        1 SMS: 1 sen
        1 GB data: USD7

### Memperbaiki Data

Terdapat beberapa kesalahan umum pada data, terutama tipe data tanggal yang masih bertipe string/ object. Di sisi lain, untuk kepentingan analisis ini, kita sebenarnya hanya membutuhkan data bulanan.

Oleh karena itu, kami akan mengabaikan tipe data pada kolom-kolom yang memuat data tanggal, tetapi kami menambah kolom baru bertipe datetime yang hanya mengambil periode berdasarkan bulan. Untuk itu, kami menggunakan fungsi yang didefinisikan secara khusus.

```python
def to_month(dataset, col_name):
    temp_array = pd.to_datetime(dataset[col_name])
    temp_array = temp_array.dt.to_period("M")
    return temp_array
```

```python
calls['month'] = to_month(calls, 'call_date')
calls.info()
calls.head()
```

Fungsi berjalan dengan baik. Kami akan lakukan ini untuk data lainnya.

```python
messages['month'] = to_month(messages, 'message_date')
messages.info()
messages.head()
```

**Pendapatan Per Paket**

Untuk mengetahui pendapatan per paket, kita perlu terlebih dahulu menemukan adanya biaya ekstra yang dibayarkan pengguna karena kelebihan penggunaan. Nilainya akan ditambahkan dengan biaya regular.

```python
dataset.head()
```

```python
# Menghitung biaya ekstra
def extra_price (value, max_value, cost_extra):
    extra_value = value - max_value
    if extra_value > 0:
        return cost_extra * extra_value
    else:
        return 0
```

```python
# Menambah kolom baru untuk menghitung total biaya
dataset['extra_internet'] = dataset.apply(lambda row: extra_price(value=row['gb_used'], max_value=row['gb_per_month_included'], cost_extra=row['usd_per_gb']), axis=1)
dataset['extra_messages'] = dataset.apply(lambda row: extra_price(value=row['sms'], max_value=row['messages_included'], cost_extra=row['usd_per_message']), axis=1)
dataset['extra_calls'] = dataset.apply(lambda row: extra_price(value=row['duration'], max_value=row['minutes_included'], cost_extra=row['usd_per_minute']), axis=1)
dataset['total_extra_cost'] = dataset[['extra_internet', 'extra_messages', 'extra_calls' ]].sum(axis=1)
dataset['total_cost'] = dataset[['total_extra_cost', 'usd_monthly_pay']].sum(axis=1)
dataset.head(50)
```

```python
income_user = dataset.groupby(['plan', 'month']).count()['user_id'].reset_index()
income_user
```

Jumlah pengguna terpantau meningkat tiap bulan.

```python
# Mencari tahu pendapatan per bulan tiap paket
income = dataset.groupby(['plan', 'month']).sum()[['usd_monthly_pay', 'total_extra_cost', 'total_cost']].reset_index()
income['month'] = income['month'].astype(str)
income
```

```python
# Perbandingan pendapatan surf vs ultimate
plt.figure(figsize=(12, 18))

plt.subplot(3, 1, 1)
plt.title('Regular Cost: Surf vs Ultimate')
sns.lineplot(data=income, x='month', y='usd_monthly_pay', hue='plan', style="plan", markers=["o", "o"])

plt.subplot(3, 1, 2)
plt.title('Extra Cost: Surf vs Ultimate')
sns.lineplot(data=income, x='month', y='total_extra_cost', hue='plan', style="plan", markers=["o", "o"])

plt.subplot(3, 1, 3)
plt.title('Total Cost: Surf vs Ultimate')
sns.lineplot(data=income, x='month', y='total_cost', hue='plan', style="plan", markers=["o", "o"])

plt.show()
```

**Findings:**
- Jumlah user pada masing-masing paket terus meningkat dari bulan ke bulan. Hal ini berimbas pada naiknya pendapatan pada masing-masing paket.
- Jumlah pengguna paket surf pada akhir tahun (Desember) mencapai dua kali lipat jumlah pengguna paket ultimate (303 vs 146).
- Naiknya pendapatan regular pada masing-masing paket sejalan dengan naiknya jumlah user per bulan.
- Biaya ekstra dari pelanggan surf jauh lebih tinggi ketimbang biaya esktra dari pelanggan ultimate. Bahkan, pelanggan ultimate cenderung tidak terkena biaya ekstra.
- Tingginya biaya ekstra pada paket surf menyebabkan total biaya yang dibayar oleh pelanggan paket surf pada akhirnya justru lebih tinggi ketimbang paket ultimate.
- Biaya ekstra yang dikeluarkan oleh pelanggan surf justru melampaui biaya regular dari paket surf itu sendiri, bahkan juga melampaui biaya regular dari pelanggan ultimate.

**Insights:**
- Pelanggan cenderung memilih paket yang lebih murah (surf), padahal tingkat konsumsi paket prabayar mereka cenderung lebih tinggi dibanding yang disediakan oleh paket tersebut.
- Pelanggan cenderung tidak keberatan menggeluarkan biaya ekstra untuk pembelian paket yang lebih murah ketimbang sejak awal membeli paket yang lebih mahal yang menyediakan layanan lebih banyak.

## Uji Hipotesis

Kita sudah menyelesaikan proses analisis data, kini saatnya untuk menguji hipotesis awal.

- Rata-rata pendapatan dari pengguna paket telepon Ultimate dan Surf berbeda.
- Rata-rata pendapatan dari pengguna di wilayah NY-NJ berbeda dengan pendapatan pengguna dari wilayah lain.


### Tes Hipotesis 1

*Rata-rata pendapatan dari pengguna paket telepon Ultimate dan Surf berbeda.*

Sebelumnya, kami sudah menemukan bahwa nilai rata-rata pendapatan kedua paket ini *memang* berbeda, yang mana pendapatan surf lebih tinggi ketimbang ultimate. Namun, kami perlu mempertanggungjawabkan ini dengan metode statistik yang dapat diterima. 

- Hipotesis nol **(H0)** kita yakni 'Rata-rata pendapatan dari pengguna paket telepon Ultimate dan Surf **berbeda.**' 
- Dengan demikian, hipotesis alternatifnya **(H1)** menjadi 'Rata-rata pendapatan dari pengguna paket telepon Ultimate dan Surf **sama**.' 

Kami menggunakan metode **st.ttest_ind()**, yakni metode yang digunakan untuk menguji hipotesis persamaan rata-rata dua populasi statistik yang berbeda. Dua populasi yang akan kita uji yakni *dataset_surf* dan *dataset_ultimate*.

Kami menggunakan nilai alpha 5% sebagai ambang batas. Jika nilai probabilitas (pvalue) yang kami dapatkan kurang dari 5%, itu berarti rata-rata kedua populasi ini tidak sama, sehingga H0 bisa diterima. Sebaliknya, jika pvalue berada di atas 5%, kita harus menolak H0.

Kami juga akan memeriksa terlebih dahulu, apakah varians kedua data ini sama.

```python
dataset_surf['total_cost'].var()
```

```python
dataset_ultimate['total_cost'].var()
```

Kedua dataset ini memiliki varian yang berbeda, sehingga equal_var akan kita atur sebagai False.

```python
alpha = 0.05
results = st.ttest_ind(dataset_surf['total_cost'], dataset_ultimate['total_cost'], equal_var=False)
print('p-value:', results.pvalue)
if (results.pvalue < alpha):
    print("Kita menerima hipotesis nol")
else:
    print("Kita menolak hipotesis nol")
```

Berdasarkan hasil pengujian hipotesis ini, kami memutuskan untuk menerima hipotesis bahwa:\
**'Rata-rata pendapatan dari pengguna paket telepon Ultimate dan Surf berbeda.'**

<div class="alert alert-block alert-warning">
<b>Reviewer's comment v1</b> <a class="tocSkip"></a>

Secara garis besar, sebenarnya yang kamu lakukan sudah benar, tetapi umumnya uji hipotesis menggunakan H0 sebagai hipotesis yang akan **ditolak** jika `p-value < alpha`. Jadi, antara H0 dan H1 kamu terbalik, jika mengikuti konvensi umum, tetapi cara uji hipotesis yang kamu lakukan masih benar.
</div>

### Tes Hipotesis 2

*Rata-rata pendapatan dari pengguna di wilayah NY-NJ berbeda dengan pendapatan pengguna dari wilayah lain.*

Kami akan melakukan uji hipotesis seperti yang pertama dengan menggunakan *dataset_city*. 

- H0 : *Rata-rata pendapatan dari pengguna di wilayah NY-NJ **berbeda** dengan pendapatan pengguna dari wilayah lain.*
- H1 : *Rata-rata pendapatan dari pengguna di wilayah NY-NJ **sama** dengan pendapatan pengguna dari wilayah lain.*

```python
def city_code_new(city):
    if 'NY' in city:
        return 'NY-NJ'
    else:
        return 'Others'
```

Kami sudah memiliki kolom baru yang mengelompokkan kota menjadi dua, yakni NY-NJ dan Others. Kami akan membagi ini menjadi dua kelompok dataset untuk diuji bersama-sama nantinya.

Berdasarkan hasil pengujian hipotesis ini, kami memutuskan untuk menolak hipotesis nol. Dengan demikian, kami menerima hipotesis alternatifnya (H1) yakni bahwa:\
